```python
import matplotlib.pyplot as plt
import numpy as np

def derivative(x,y):
    return np.array([x ** 3 - 8 * x - 3, y ** 3 - 8 * y - 3])

def f(x,y):
    return 1/4 * x ** 4 + 1/4 * y ** 4 - 4 * x ** 2 - 4 * y ** 2 - 3 * x - 3 * y
x = np.linspace(-4,4,100)
y = np.linspace(-4,4,100)

xx, yy = np.meshgrid(x,y)
z = 1/4 * xx ** 4 + 1/4 * yy ** 4 - 4 * xx ** 2 - 4 * yy ** 2 - 3 * xx - 3 * yy

plt.contourf(x,y,z,20)
plt.axis('scaled')
plt.show()
```

```python
x = np.linspace(-5.5,5.5,100)
y = np.linspace(-5.5,5.5,100)

xx, yy = np.meshgrid(x,y)
z = 1/4 * xx ** 4 + 1/4 * yy ** 4 - 4 * xx ** 2 - 4 * yy ** 2 - 3 * xx - 3 * yy

plt.contourf(x,y,z,30)
plt.axis('scaled')
plt.show()

starting_points = [(-5,-5), (-5,-2), (4,4)]
for point in starting_points:
    # point = (np.random.rand(2) - 0.5) * 8
    print("initial point")
    print(point)
    a = 0.01
    points = [point]

    for i in range(100):
        grad = derivative(*point)
        value = f(*point)
        if i % 10 == 0:
            print(f"value: {value}, gradient: {grad}")

        point = point - a * grad
        points.append(point)

    plt.contourf(x,y,z,20)
    plt.axis('scaled')
    plt.plot(*zip(*points), c="r")
```
